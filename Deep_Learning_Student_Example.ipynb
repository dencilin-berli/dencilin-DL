# Deep Learning Example for Students

Predict Pass or Fail based on Study Hours

## Step 1: Import Libraries

```
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Step 2: Create Dataset

```
# Input data (study hours)
X = np.array([1, 2, 3, 4, 5, 6])

# Output data (0 = Fail, 1 = Pass)
y = np.array([0, 0, 0, 1, 1, 1])
```

## Step 3: Build Deep Learning Model

```
model = Sequential()
model.add(Dense(4, input_shape=(1,), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## Step 4: Compile Model

```
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

## Step 5: Train Model

```
model.fit(X, y, epochs=200, verbose=0)
```

## Step 6: Test Model

```
prediction = model.predict(np.array([3.5]))
print(prediction)

if prediction > 0.5:
    print('Pass')
else:
    print('Fail')
    sql -->ble  js
```
